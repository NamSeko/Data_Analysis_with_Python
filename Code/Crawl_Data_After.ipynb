```python
# import c√°c th∆∞ vi·ªán c·∫ßn s·ª≠ d·ª•ng ƒë·ªÉ crawl d·ªØ li·ªáu
from facebook_scraper import get_posts
import pandas as pd
import numpy as np
```

```python
FANPAGE_LINK ="runyouredmxyz"
FOLDER_PATH = "../Data/"
COOKIE_PATH = "../Data/www.facebook.com_cookies.txt"

PAGES_NUMBER = 10 # Number of pages to crawl
```

```python
# Crawl d·ªØ li·ªáu t·ª´ trang v·ªÅ sau ƒë√≥ th√™m v√†o post_list
end_post_id = 714164637406487
post_list = []
for post in get_posts(FANPAGE_LINK,
                    options={"comments": True, "reactions": True, "allow_extra_requests": True},
                    extra_info=True, pages=PAGES_NUMBER, cookies=COOKIE_PATH):
    print(post)
    if post['post_id'] == end_post_id:
        break
    post_list.append(post)
```

```python
# Kh·ªüi t·∫°o DataFrame
data = pd.DataFrame(columns=post_list[0].keys(), index=range(len(post_list)), data=post_list)

FILE_NAME = FOLDER_PATH + 'Data_After'
# Chuy·ªÉn d·ªØ li·ªáu sang file csv, xlsx,... ƒë·ªÉ d·∫Ω thao t√°c v·ªõi d·ªØ li·ªáu
path1=FILE_NAME + ".csv"
data.to_csv(path1, index=False)
print(path1)
```
